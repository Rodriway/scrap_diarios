```python
import pandas as pd
from google.colab import files
import pandas as pd
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException

from datetime import datetime
import pytz
pd.set_option('display.max_colwidth', 20)
from newspaper import Article

# Subir archivos Parquet desde tu ordenador local a Colab
uploaded = files.upload()

# Obtener la lista de nombres de los archivos subidos
file_names = list(uploaded.keys())

# Leer archivos Parquet y almacenar los DataFrames resultantes en una lista
dfs = [pd.read_parquet(file_name) for file_name in file_names]

# Concatenar los DataFrames en uno solo
concatenated_df = pd.concat(dfs, ignore_index=True)
concatenated_df.drop(columns="titulo",inplace=True)

df_unicos = concatenated_df.drop_duplicates(subset=['link'], keep='last')

# Crear columnas vacías en df_unicos
df_unicos["titulo"] = ""
df_unicos["autor"] = ""
df_unicos["fecha_publi"] = ""
df_unicos["texto"] = ""
```

```python
# Configuración del navegador
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--disable-dev-shm-usage')


# Obtener el número de filas con valores vacíos en las cuatro columnas
numero_filas_vacias_en_4_columnas = df_unicos[df_unicos["titulo"].eq("") | df_unicos["autor"].eq("") | df_unicos["fecha_publi"].eq("") | df_unicos["texto"].eq("")].shape[0]

# Imprimir mensaje de inicio
print("Se comienza con la búsqueda de", numero_filas_vacias_en_4_columnas, "artículos")

# Inicializar el navegador
driver = webdriver.Chrome(options=chrome_options)

# Iterar sobre los enlaces únicos
for index, row in df_unicos.iterrows():
    link = row["link"]

    # Verificar si las cuatro columnas están vacías para este artículo
    if row["titulo"] == "" or row["autor"] == "" or row["fecha_publi"] == "" or row["texto"] == "":
        print("Buscando información para el artículo:", link)

        try:
            # Ingresar a la URL
            driver.get(link)
            time.sleep(2)  # Esperar un momento a que la página cargue

            # Extraer el título de la noticia
            titulo = driver.find_element(By.CLASS_NAME, "articulo__titulo").text

            # Extraer el autor
            autor_element = driver.find_element(By.CSS_SELECTOR, ".articulo__opinologo-item")
            autor = autor_element.text.strip()

            # Intentar extraer la fecha de publicación
            try:
                fecha_element = driver.find_element(By.CSS_SELECTOR, ".articulo__fecha")
                fecha_publi = fecha_element.text.strip()
            except NoSuchElementException:
                fecha_publi = None

            # Intentar extraer el texto completo usando Newspaper3k
            try:
                article = Article(link)
                article.download()
                article.parse()
                texto = article.text.strip()
            except:
                texto = None

            # Actualizar el DataFrame con la información obtenida
            df_unicos.at[index, "titulo"] = titulo
            df_unicos.at[index, "autor"] = autor
            df_unicos.at[index, "fecha_publi"] = fecha_publi
            df_unicos.at[index, "texto"] = texto

        except Exception as e:
            # Imprimir mensaje de error
            print("Ocurrió un error:", str(e))
            print("Reiniciando desde el índice:", index)
            break  # Salir del bucle

# Cerrar el navegador
driver.quit()


df_final=pd.merge(concatenated_df,df_unicos[["link","titulo","autor","fecha_publi","texto"]], on="link")
df_final.shape, concatenated_df.shape, df_unicos.shape
df_final['hora_ejecucion'] = df_final['hora_ejecucion'].apply(lambda x: datetime.strptime(x, "%I %p").strftime("%H:%M:%S"))

# Nombre del archivo
nombre_archivo = f"scrap_lanueva_{df_final['fecha'].min()}_a_{df_final['fecha'].max()}.parquet"
nombre_archivo

# Guardar el DataFrame como archivo Parquet
df_final.to_parquet(nombre_archivo)

from google.colab import files
files.download(nombre_archivo)
```
